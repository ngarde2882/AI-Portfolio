```python
#######################################
# Minmax search
#
# Yoonsuck Choe
# 9/26/2021 - revised 9/28
#
# - This is a really basic implementation.
# - You can consider modifying it to keep track of the solution path
#
#######################################

#######################################
# Constants

MAX = 99999
MIN = -99999

#######################################
# Game tree  : Change the list "Game_tree" to test a different game
'''
                       o MAX
                       |
           ____________|_________
          |                      |
    ______o________           ___o___     MIN
   |      |        |         |       |
   7    __o___    _o_        12    __o__       MAX
       |  |  |   |   |            |     |
       3  8  1   15  9           17     13
'''

Game_tree = [[7, [3, 8, 1], [15, 9]], [12, [17, 13]]]
HW_tree = [[[5,8],[3,[5,2,7,0],9,[5,8,1]],[7,[12,10,22]]],[[[3,6,4],6,[10,3]],[[8,13],5,[10,8]]]]
#######################################
# Function definitions

def minmax_decision(tree):
  '''
  input: game tree (as a python list)
  return: value of optimal outcome
  '''

  # long implementation (as in the slides)
  '''
  max = MIN

  for i in tree:
    v = minmax_value(i,"min")
    if (v > max):
      max = v

  return max
  '''

  # compact implementation:

  return minmax_value(tree,"max")


def minmax_value(tree,node_type,indent="",alpha=MIN,beta=MAX,path=[]):
  '''
  input: game tree and the type of its root : "max" or "min"
  return: value of optimal outcome
  '''
  print(indent+"a:",alpha,"b:",beta)
  if (type(tree)==int):      # Leaf node
    print(indent+"CALL",node_type,"returns ",tree)
    return tree

  elif (node_type == "max"): # "tree"'s root is a MAX node

    max_v = MIN

    print(indent+"CALL max "+str(tree))

    for index, i in enumerate(tree):
      v = minmax_value(i,"min",indent+"  ",alpha,beta)
      if (v>max_v):
        max_v = v
      alpha = max(alpha,v)
      if v>=beta:
        if alpha>MIN:
          print(f'{indent}   - max {tree} returns  {v} and B-cuts ({alpha}>={beta})')
          if index==len(tree)-1:
            print(f'{indent}  but there was nothing left to cut') # could put this above print and break instead
          return v


    print(indent+"  - max",tree,"returns "+str(max_v))
    return max_v

  elif (node_type == "min"): # "tree"'s root is a MIN node

    min_v = MAX

    print(indent+"CALL min "+str(tree))

    for index, i in enumerate(tree):
      v = minmax_value(i,"max",indent+"  ",alpha,beta)
      if (v<min_v):
        min_v = v
      beta = min(beta,v)
      if v<=alpha:
        if beta<MAX:
          print(f'{indent}   - min {tree} returns  {v} and a-cuts ({alpha}>={beta})')
          if index==len(tree)-1:
            print(f'{indent}  but there was nothing left to cut') # could put this above print and break instead
          return v

    print(indent+"  - min",tree,"returns "+str(min_v))
    return min_v

#######################################
# Test

minmax_decision(HW_tree)
```
